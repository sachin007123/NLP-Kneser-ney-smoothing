```python
def min_edit_distance(s1, s2):
    m, n = len(s1), len(s2)

    # Create DP table
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Initialize first row and first column
    for i in range(m + 1):
        dp[i][0] = i

    for j in range(n + 1):
        dp[0][j] = j

    # Fill DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(
                    dp[i][j - 1],      # Insert
                    dp[i - 1][j],      # Delete
                    dp[i - 1][j - 1]   # Replace
                )

    return dp[m][n], dp


def print_table(s1, s2, dp):
    print("\nDP Table:")
    print("    ", end="")
    for ch in " " + s2:
        print(f"{ch:>4}", end="")
    print()

    for i in range(len(dp)):
        if i == 0:
            print(" ", end="")
        else:
            print(s1[i - 1], end="")
        print(" ", end="")

        for value in dp[i]:
            print(f"{value:>4}", end="")
        print()


def backtrack_operations(s1, s2, dp):
    i = len(s1)
    j = len(s2)
    operations = []

    while i > 0 or j > 0:
        if i > 0 and j > 0 and s1[i - 1] == s2[j - 1]:
            i -= 1
            j -= 1

        elif i > 0 and j > 0 and dp[i][j] == dp[i - 1][j - 1] + 1:
            operations.append(
                f"Replace '{s1[i-1]}' at position {i-1} with '{s2[j-1]}'"
            )
            i -= 1
            j -= 1

        elif i > 0 and dp[i][j] == dp[i - 1][j] + 1:
            operations.append(
                f"Delete '{s1[i-1]}' from position {i-1}"
            )
            i -= 1

        else:
            operations.append(
                f"Insert '{s2[j-1]}' at position {i}"
            )
            j -= 1

    operations.reverse()
    return operations


# Main Program
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

distance, dp = min_edit_distance(s1, s2)

print("\nMinimum Edit Distance =", distance)

print_table(s1, s2, dp)

print("\nOperations:")
ops = backtrack_operations(s1, s2, dp)

if ops:
    for op in ops:
        print("-", op)
else:
    print("No operations required. Strings are identical.")
```
